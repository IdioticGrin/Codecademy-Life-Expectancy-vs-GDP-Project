```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('all_data.csv')
print(data.head())
print(data.info())
print(data.Country.unique())
print(data.Year.unique())

sns.set_style('darkgrid')
sns.set_palette('Paired')

for j in data.Country.unique():
    plt.subplots(1,2, figsize = (15,5))
    ax1 = plt.subplot(1,2,1)
    ax2 = plt.subplot(1,2,2)
    ax1 = sns.lineplot(ax=ax1, x=data.Year, y=data['Life expectancy at birth (years)'][data.Country == j])
    ax1.set_title('Life Expectancy of ' + j + ' by Year', fontsize = 14)
    ax1.set_xticks(data.Year)
    ax1.set_xticklabels(data.Year, rotation = 45, fontsize = 13)
    ax2 = sns.lineplot(ax=ax2, x=data.Year, y=data.GDP[data.Country == j])
    ax2.set_title('GDP of ' + j + ' per Year', fontsize = 14)
    ax2.set_xticks(data.Year)
    ax2.set_xticklabels(data.Year, rotation = 45, fontsize = 13)
    
    plt.savefig(j + '.png')
    plt.show()


#for i in data.Country.unique():
    #sns.lineplot(x=data.Year, y=data['Life expectancy at birth (years)'][data.Country == i])
    #plt.title('Life Expectancy of ' + i + ' by Year')
    #plt.show()
    
#for i in data.Country.unique():
    #sns.lineplot(x=data.Year, y=data.GDP[data.Country == i])
    #plt.title('GDP of ' + i + ' per Year')
    #plt.show()



plt.subplots(1,2, figsize=(15,5))
ax1=plt.subplot(1,2,1)

ax1 = sns.lineplot(x=data.Year, y=data['Life expectancy at birth (years)'], hue=data.Country)
ax1.set_title('Life Expectancy of All Six Countries')


ax2=plt.subplot(1,2,2)
ax2 = sns.lineplot(x=data.Year, y=data.GDP, hue=data.Country)
ax2.set_title('GDP of all Six Countries')

plt.savefig('all_6.png')
plt.show()
```
